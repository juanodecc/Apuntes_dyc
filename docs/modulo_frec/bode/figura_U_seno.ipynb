```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np
```

```python
G=ctrl.tf(1,[1, 1])
G
```

para una se√±al de entrada sinusoidal tenemos:

```python
w0=.5
t=np.linspace(0,20,1001)
u=np.sin(w0*t)
t,y=ctrl.forced_response(G,T=t, U=u)
plt.figure()
plt.plot(t, y, 'b',t,u,'grey')
plt.grid()
```

En la siguiente figura se muestra el resultado para entradas sinusoidales con distintos valores de $\omega_0$

```python
fig, axs = plt.subplots(3,2, figsize=(15, 10), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .7, wspace=.2)

axs = axs.ravel()
w0=np.array([.1,.5,1,2,3,5])
for i in range(6):
    if i==0:
        t=np.linspace(0,200,1001)
    else:
        t=np.linspace(0,20,1001)
        
    u=np.sin(w0[i]*t)
    t,y=ctrl.forced_response(G,T=t, U=u)
    
    axs[i].plot(t, y, 'b',t,u,'grey')
    axs[i].set_title(r'$\omega_0$ = '+str(w0[i]))
    axs[i].set_xlabel('Tiempo [s]')
    axs[i].set_ylabel('Amplitud')
    axs[i].grid()
```

```python
ctrl.bode(G,dB=True);
```
