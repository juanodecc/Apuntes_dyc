```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import numpy as np
#import control as ctrl

x1 = np.empty((60,2))
y1 = np.empty((60,2))
for kp in range(0,60):
    den = [1,10,kp]
    s = np.roots(den)
    x1[kp] = np.real(s)
    y1[kp] = np.imag(s)

fig,ax = plt.subplots()

def animate(i):
    ax.clear()
    ax.grid()
    ax.set_xlim(-11,1)
    ax.set_ylim(-4,4)
    lc_pole=ax.plot(x1[0:i,0],y1[0:i,0],'b*',x1[0:i,1],y1[0:i,1],'g*',x1[0],y1[0],'rx')
    return lc_pole
        
ani = FuncAnimation(fig, animate, interval=60, blit=True, repeat=True, frames=100)    
ani.save("rlocus_2ord.gif", dpi=300, writer=PillowWriter(fps=25))
```
