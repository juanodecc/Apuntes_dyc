```python
import control as ctrl
import matplotlib.pyplot as plt
```

```python
A=10
B=50
Tao1=1/60
Tao2=1/600

Guy=ctrl.tf(A,[Tao1 , 1])*ctrl.tf(1,[Tao2 , 1])
Guy
```

```python
Gdy=ctrl.tf(B,[Tao1 , 1])*ctrl.tf(1,[Tao2 , 1])
Gdy
```

```python
kla=1/10
```

```python
#a lazo abierto
t,y_rla=ctrl.step_response(100*kla*Guy)
plt.plot(t,y_rla)
plt.grid()
y_rla[-1]
```

```python
#a lazo cerrado
klc=9.9 # esto es para lograr un rechazo de perturbaciones tal que 1+A*klc=100

Try_lc=ctrl.feedback(klc*Guy)
```

```python
t,y_rlc=ctrl.step_response(100*Try_lc)
plt.plot(t,y_rlc)
plt.grid()
y_rlc[-1]
```
